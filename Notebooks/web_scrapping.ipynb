```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import re
import pandas as pd
from selenium.common.exceptions import NoSuchElementException

PATH = 'D:\personal\Personal Projects\salary_predictor\chromedriver\chromedriver.exe'

target_url = "https://www.glassdoor.co.in/Job/pune-india-data-scientist-jobs-SRCH_IL.0,10_IC2856202_KO11,25.htm"

driver=webdriver.Chrome(PATH)

driver.get(target_url)


# Click "Show more jobs" button as long as it is available
while True:
    try:
        # Find and click the "Show more jobs" button
        show_more_button = driver.find_element(by=By.XPATH, value='//button[@data-test="load-more"]')
        show_more_button.click()
        time.sleep(2)  # Wait for new jobs to load
    except NoSuchElementException:
        print("No more jobs to load.")
        break  # Break the loop if the button is not found

job_listings = driver.find_elements(by = By.CLASS_NAME, value = 'JobCard_jobCardContainer___hKKI')


# Initialize a list to store job data
job_data = []


for i in range(len(job_listings)):

    job_listings[i].click()
    time.sleep(1)
    
    # Extracting job details
    try:
        # Try to extract the rating
        rating = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[1]/div[4]/div[2]/div[2]/div/div[1]/header/div[1]/a/div[2]/div[2]').text
    except:
        rating = 'N/A'  # Set to 'N/A' if not found

    try:
        # Extract the employer name
        employer = driver.find_element(by=By.XPATH, value='//h4[contains(@class, "heading_Subhead__Ip1aW")]').text
    except:
        employer = 'N/A'  # Set to 'N/A' if not found

    try:
        # Extract the job title
        title = driver.find_element(by=By.XPATH, value='//h1[@class="heading_Heading__BqX5J heading_Level1__soLZs"]').text

    except:
        title='N/A'
    
    try:
        # Extract the location
        location = driver.find_element(by=By.XPATH, value='//div[@class="JobDetails_location__mSg5h" and @data-test="location"]').text

    except:
        location='N/A'

    try:
        description = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[1]/div[4]/div[2]/div[2]/div/div[1]/section/div[2]/div[1]').text
    except:
        description='N/A'
    try:
        salary_estimate = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[1]/div[4]/div[2]/div[2]/div/div[1]/section/section[1]/div/div[1]/div[1]/div[2]').text
    except:
        salary_estimate = 'N/A'

    try:
        company_overview = driver.find_element(by=By.XPATH, value='/html/body/div[3]/div[1]/div[4]/div[2]/div[2]/div/div[1]/section/section[2]').text
    except:
        company_overview = 'N/A'

    # Append the job details to the list
    job_data.append({
        "Employer": employer,
        "Rating": rating,
        "Job Title": title,
        "Location": location,
        "Description": description,
        "Salary Estimate": salary_estimate,
        "Company Overview": company_overview
    })

    # Print the results for the current job post
    print(f"Employer: {employer}")
    print(f"Rating: {rating}")
    print(f"Job Title: {title}")
    print(f"Location: {location}")
    print(f"Description: {description}")
    print(f"Salary Estimate: {salary_estimate}")
    print(f"Company Overview: {company_overview}")
    print()

# Create a DataFrame from the job data
job_df = pd.DataFrame(job_data)

# Export the DataFrame to an Excel file
job_df.to_excel("job_listings.xlsx", index=False)
```
